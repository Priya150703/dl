```python
#example of using a pre-trained model as a classifier
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from keras.applications.vgg16 import preprocess_input
from keras.applications.vgg16 import decode_predictions
from keras.applications.vgg16 import VGG16
import matplotlib.pyplot as plt

from google.colab import files

uploaded = files.upload()

image = load_img('download2.jpg', target_size=(224, 224)) # load image from file
image_array = img_to_array(image) # convert the image pixels to a numpy array
image_array = image_array.reshape((1, image_array.shape[0], image_array.shape[1]
                                   , image_array.shape[2]))
# reshape data for model input shape (224x224 pixels with 3 color channels)
image_array = preprocess_input(image_array) # prepare image for the VGG model
model = VGG16() # load the model
yhat = model.predict(image_array) #predict probability across all output classes
label = decode_predictions(yhat)
#converts probabilities to class labels with human-readable names
label = label[0][0] # retrieve the most likely result, e.g. highest probability
print('%s (%.2f%%)' % (label[1], label[2]*100)) # print the classification
plt.imshow(image)
plt.axis('off')  # Turn off x and y axes
plt.show()
```
